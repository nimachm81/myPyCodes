```python
%reload_ext autoreload
%autoreload
%pylab inline

from scipy import constants
import numpy as np
import math

import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

from Electromagnetics.graphene import *

mu_c = np.linspace(0.001, 0.2, 5)*constants.eV
B_0 = np.linspace(0.0, 0.1, 5)
tau = 0.1*constants.pico
T = 300.0

freq = 0.1*constants.tera
omega_0 = 2.0*math.pi*freq

X, Y = np.meshgrid(mu_c/constants.eV, B_0)
Z = np.zeros(X.shape)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        sigma_d, sigma_o = condKuboLorentzian(mu_c[j], B_0[i], tau, np.array([omega_0]), T)
        Z[i, j] = abs(sigma_o[0]/sigma_d[0])
        

figsize(12, 8)

im = plt.imshow(Z, interpolation='bilinear', origin='lower', extent=[mu_c[0]/constants.eV, mu_c[-1]/constants.eV, B_0[0], B_0[-1]], cmap=cm.rainbow)
CBI = plt.colorbar(im, orientation='vertical', shrink=0.8)


CS = plt.contour(X, Y, Z)
plt.clabel(CS, inline=1, fontsize=10)
#CB = plt.colorbar(CS, shrink=0.8, extend='both')

xlabel('$\\mu_c$ - (eV)')
ylabel('$B_0$ - (T)')
plt.title('$|\\sigma_o/\\sigma_d|$ - $f={}$ THz, $\\tau={}$ ps'.format(freq/constants.tera, tau/constants.pico))

font = {'family' : 'serif', 'weight' : 'normal', 'size'   : 18}
matplotlib.rc('font', **font)


import os
cwd = os.getcwd()
dir_name = os.path.join(cwd, 'other', 'graphene-magnetic-sensor', 'sigo-sigd')
file_ext = '.png'
if not os.path.exists(dir_name):
    os.makedirs(dir_name)

file_name = os.path.join(dir_name, 'mu-B0'+file_ext)
savefig(file_name)


plt.show()
```

### broadside coupled graphene strips ( n-p doping )

```python
%reload_ext autoreload
%autoreload 2

import numpy as np 
from Electromagnetics.graphene import *
from scipy import constants

from Electromagnetics.MaxwellEigState2D import *
from Electromagnetics.Misc import *
import math

del_graph = 50.0*constants.nano
width_graph = 10.0*constants.micro
gap_graph = 1.0*constants.micro

W_x = 6.0*width_graph
W_y = gap_graph + 4.0*width_graph


fs_name = 'freespace'
fs_rect = [Point2D(0.0, 0.0), Point2D(W_x, W_y)]
fs_type = 'all_scalar'
def fs_eps(f):
    return constants.epsilon_0
def fs_mu(f):
    return constants.mu_0
def fs_sig(f):
    return 0.0
fs_params = [fs_eps, fs_mu, fs_sig]
fs_mat = [fs_name, fs_rect, fs_type, fs_params]


P00 = Point2D(W_x/2.0-width_graph/2.0, W_y/2.0-gap_graph/2.0-del_graph/2.0)
P11 = Point2D(W_x/2.0+width_graph/2.0, W_y/2.0-gap_graph/2.0+del_graph/2.0)
graph_name = 'graphene-1'
graph_rect = [P00, P11]
graph_type = 'sigma_tensor'
def graph_eps(f):
    return constants.epsilon_0
def graph_mu(f):
    return constants.mu_0

mu_c = 0.02*constants.eV
B_0 = 0.1
tau = 0.2*constants.pico
T = 100.0
def graph_sig_1(f):
    omega_0 = 2.0*math.pi*f
    sigma_d, sigma_o = condKuboLorentzian(mu_c, B_0, tau, np.array([omega_0]), T)
    sig_d, sig_o = sigma_d[0]/del_graph, sigma_o[0]/del_graph    ## n-doped
    print('sig_o/sig_d: ', abs(sig_o/sig_d))
    sig = [[sig_d , 0.0  ,  sig_o],
           [0.0   , 0.0,  0.0  ],
           [-sig_o, 0.0  ,  sig_d]]
    return sig
graph_params = [graph_eps, graph_mu, graph_sig_1]
graph_mat_1 = [graph_name, graph_rect, graph_type, graph_params]


P00 = Point2D(W_x/2.0-width_graph/2.0, W_y/2.0+gap_graph/2.0-del_graph/2.0)
P11 = Point2D(W_x/2.0+width_graph/2.0, W_y/2.0+gap_graph/2.0+del_graph/2.0)
graph_name = 'graphene-2'
graph_rect = [P00, P11]

def graph_sig_2(f):
    omega_0 = 2.0*math.pi*f
    sigma_d, sigma_o = condKuboLorentzian(mu_c, B_0, tau, np.array([omega_0]), T)
    sig_d, sig_o = sigma_d[0]/del_graph, -sigma_o[0]/del_graph    ## p-doped
    sig = [[sig_d , 0.0  ,  sig_o],
           [0.0   , 0.0,  0.0  ],
           [-sig_o, 0.0  ,  sig_d]]
    return sig
graph_params = [graph_eps, graph_mu, graph_sig_2]
graph_mat_2 = [graph_name, graph_rect, graph_type, graph_params]

w_edge_mode = 3.0*constants.micro
P00 = Point2D(W_x/2.0-width_graph/2.0-w_edge_mode, W_y/2.0-w_edge_mode)
P11 = Point2D(W_x/2.0-width_graph/2.0+w_edge_mode, W_y/2.0+w_edge_mode)
rect_edge_l = [P00, P11]
P00 = Point2D(W_x/2.0+width_graph/2.0-w_edge_mode, W_y/2.0-w_edge_mode)
P11 = Point2D(W_x/2.0+width_graph/2.0+w_edge_mode, W_y/2.0+w_edge_mode)
rect_edge_r = [P00, P11]
w_bulk_mode = 5.0*constants.micro
P00 = Point2D(W_x/2.0-width_graph/2.0-w_bulk_mode, W_y/2.0-w_bulk_mode)
P11 = Point2D(W_x/2.0+width_graph/2.0+w_bulk_mode, W_y/2.0+w_bulk_mode)
rect_bulk = [P00, P11]
mesh_params = [[rect_bulk, 3], [rect_edge_l, 2], [rect_edge_r, 2]] 


maxwell2Dsolver = MaxEigState2DSolver([fs_mat, graph_mat_1, graph_mat_2], mesh_params, vb=True)

firstPass = True
```

### in-plane coupled graphene strips ( n-n doping )

```python
%reload_ext autoreload
%autoreload 2

import numpy as np 
from Electromagnetics.graphene import *
from scipy import constants

from Electromagnetics.MaxwellEigState2D import *
from Electromagnetics.Misc import *
import math

del_graph = 50.0*constants.nano
width_graph = 10.0*constants.micro
gap_graph = 1.0*constants.micro

W_x = gap_graph + 6.0*width_graph
W_y = 4.0*width_graph


fs_name = 'freespace'
fs_rect = [Point2D(0.0, 0.0), Point2D(W_x, W_y)]
fs_type = 'all_scalar'
def fs_eps(f):
    return constants.epsilon_0
def fs_mu(f):
    return constants.mu_0
def fs_sig(f):
    return 0.0
fs_params = [fs_eps, fs_mu, fs_sig]
fs_mat = [fs_name, fs_rect, fs_type, fs_params]


P00 = Point2D(W_x/2.0-gap_graph/2.0-width_graph, W_y/2.0-del_graph/2.0)
P11 = Point2D(W_x/2.0-gap_graph/2.0, W_y/2.0+del_graph/2.0)
graph_name = 'graphene-1'
graph_rect = [P00, P11]
graph_type = 'sigma_tensor'
def graph_eps(f):
    return constants.epsilon_0
def graph_mu(f):
    return constants.mu_0

mu_c = 0.02*constants.eV
B_0 = 0.3 #0.01
tau = 0.2*constants.pico
T = 100.0
def graph_sig_1(f):
    omega_0 = 2.0*math.pi*f
    sigma_d, sigma_o = condKuboLorentzian(mu_c, B_0, tau, np.array([omega_0]), T)
    sig_d, sig_o = sigma_d[0]/del_graph, sigma_o[0]/del_graph   ## n-doped
    print('sig_o/sig_d: ', abs(sig_o/sig_d))
    sig = [[sig_d , 0.0  ,  sig_o],
           [0.0   , 0.0,  0.0  ],
           [-sig_o, 0.0  ,  sig_d]]
    return sig
graph_params = [graph_eps, graph_mu, graph_sig_1]
graph_mat_1 = [graph_name, graph_rect, graph_type, graph_params]


P00 = Point2D(W_x/2.0+gap_graph/2.0, W_y/2.0-del_graph/2.0)
P11 = Point2D(W_x/2.0+gap_graph/2.0+width_graph, W_y/2.0+del_graph/2.0)
graph_name = 'graphene-2'
graph_rect = [P00, P11]

def graph_sig_2(f):
    omega_0 = 2.0*math.pi*f
    sigma_d, sigma_o = condKuboLorentzian(mu_c, B_0, tau, np.array([omega_0]), T)
    sig_d, sig_o = sigma_d[0]/del_graph, sigma_o[0]/del_graph    ## n-doped
    sig_d, sig_o = 0.0, 0.0
    sig = [[sig_d , 0.0  ,  sig_o],
           [0.0   , 0.0,  0.0  ],
           [-sig_o, 0.0  ,  sig_d]]
    return sig
graph_params = [graph_eps, graph_mu, graph_sig_2]
graph_mat_2 = [graph_name, graph_rect, graph_type, graph_params]

w_edge_mode = 3.0*constants.micro
P00 = Point2D(W_x/2.0-width_graph/2.0-w_edge_mode, W_y/2.0-w_edge_mode)
P11 = Point2D(W_x/2.0-width_graph/2.0+w_edge_mode, W_y/2.0+w_edge_mode)
rect_edge_l = [P00, P11]
P00 = Point2D(W_x/2.0+width_graph/2.0-w_edge_mode, W_y/2.0-w_edge_mode)
P11 = Point2D(W_x/2.0+width_graph/2.0+w_edge_mode, W_y/2.0+w_edge_mode)
rect_edge_r = [P00, P11]
w_bulk_mode = 5.0*constants.micro
P00 = Point2D(W_x/2.0-width_graph/2.0-w_bulk_mode, W_y/2.0-w_bulk_mode)
P11 = Point2D(W_x/2.0+width_graph/2.0+w_bulk_mode, W_y/2.0+w_bulk_mode)
rect_bulk = [P00, P11]
mesh_params = [[rect_bulk, 3], [rect_edge_l, 2], [rect_edge_r, 2]] 


maxwell2Dsolver = MaxEigState2DSolver([fs_mat, graph_mat_1, graph_mat_2], mesh_params, vb=True)

firstPass = True
```

#### frequency point simulation for broadside or in-plane coupled 

```python
## firstPass is set automaically to True for the first simulation attempt and to False for the consecuitive ones

freq_0 = 7.0*constants.tera
freq_1 = 1.0*constants.tera

freq = 2.0*constants.tera

from Electromagnetics.graphene import *

omega_0 = 2.0*math.pi*freq
kz_k0_inf = magnetoplasmonDispersion_normalized(mu_c, B_0, np.array([omega_0]), tau, T, cond='Kubo')

n_modes = 5
lambda_0 = kz_k0_inf[0][0]*2.1
k_eigs, k_vecs = maxwell2Dsolver.Solve(freq, n_modes, lambda_0, firstPass=firstPass)

print(k_eigs)
print(k_vecs.shape)

print('infinite magnetoplasmon dispersion graphene: \n', kz_k0_inf)

firstPass = False
```
