```python
%reload_ext autoreload
%autoreload
%pylab inline --no-import-all

import numpy as np
from scipy import constants
from Electromagnetics.Causality import *

gamma = 1.0e4
sig_0 = 1.0

def cond(omega):
    return sig_0/(1j*omega + gamma)

def cond_r(omega):
    return np.real(cond(omega))

def cond_i(omega):
    return np.imag(cond(omega))
                   
omega_0 = 1.0
omega_1 = 10.0*gamma
N = 100
omega = np.linspace(omega_0, omega_1, N)

sig_r = cond_r(omega)
sig_i = cond_i(omega)
sig_r_kk = KKIntegralR(cond_i, omega)
sig_i_kk = KKIntegralI(cond_r, omega)

figsize(15, 10)
plot(omega, sig_r, 'r')
plot(omega, sig_r_kk, 'r.')
plot(omega, sig_i, 'b')
plot(omega, sig_i_kk, 'b.')

#print(sig_r_kk)
```
