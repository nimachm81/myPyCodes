```python
import os
cwd = os.getcwd()
dir_name = os.path.join(cwd, 'other', 'test')
file_ext = '.png'
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
```

```python
%pylab inline

font = {'family' : 'serif', 'weight' : 'normal', 'size'   : 14}
matplotlib.rc('font', **font)


def getPl(v_g, z_g, z_l):
    p_l = 0.5*np.real(v_g**2*z_l/(z_g+z_l)**2)
    return p_l


v_g = 100
r_l = 50

N = 200
r_g = np.linspace(0, 100, N)
p_l = np.zeros(N)

for i in range(N):
    p_l[i] = getPl(v_g, r_g[i], r_l)
    
    
plot(r_g, p_l)
title('$V_g={}, R_L={}$'.format(v_g, r_l), fontsize=16)
xlabel('$R_g$', fontsize=16)
ylabel('$P_L$', fontsize=16)

file_name = os.path.join(dir_name, 'r-l-fixed'+file_ext)
savefig(file_name, bbox_inches='tight', pad_inches=0.5)

show()

##-----------------------------------------------------

v_g = 100
r_g = 50

N = 200
r_l = np.linspace(0, 200, N)
p_l = np.zeros(N)

for i in range(N):
    p_l[i] = getPl(v_g, r_g, r_l[i])
    
    
plot(r_l, p_l)
title('$V_g={}, R_g={}$'.format(v_g, r_g), fontsize=16)
xlabel('$R_L$', fontsize=16)
ylabel('$P_L$', fontsize=16)

file_name = os.path.join(dir_name, 'r-g-fixed'+file_ext)
savefig(file_name, bbox_inches='tight', pad_inches=0.5)

show()

```
