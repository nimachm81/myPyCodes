```python
from Finance.Finance import StockReports

sec = StockReports(form_type='8-K', form_items=['2.02'], count=100)
#sec = StockReports(form_type='10-Q', form_items=None, count=100)

feed = sec.GetSecRecetlyReportedFeed()

sec.PrintFeedInfo(feed)

indsforms2_02 = sec.FindFeeditemsHavingFormitem()
print('indsforms2_02: ', indsforms2_02)


html = sec.FeedItem_GetUrlHTMLPage(indsforms2_02[0])

txtFile = sec.HtmlPage_extractTxtFile(html)

tables = sec.TxtFile_ExtractTables(txtFile, displayTables=True)

for i in range(len(tables)):
    table = tables[i]
    print('table: ', i)
    asset_data = sec.HtmlTable_FindFinancialTermData(table, 'total assets')
    liability_data = sec.HtmlTable_FindFinancialTermData(table, 'total liabilities')
    profit_data = sec.HtmlTable_FindFinancialTermData(table, 'net profit') ## net income/profit
    income_data = sec.HtmlTable_FindFinancialTermData(table, 'net income') ## net income/profit
    loss0_data = sec.HtmlTable_FindFinancialTermData(table, 'net loss')
    loss1_data = sec.HtmlTable_FindFinancialTermData(table, 'net (loss)')
    print(asset_data, '\n', '-'*30)
    print(liability_data, '\n', '-'*30)
    print(profit_data, '\n', '-'*30)
    print(income_data, '\n', '-'*30)
    print(loss0_data, '\n', '-'*30)
    print(loss1_data, '\n', '-'*30)
    print('='*80)
    
    
```
